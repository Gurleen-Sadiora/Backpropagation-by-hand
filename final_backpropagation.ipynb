```python
#Importing libraries
import numpy as np
import pandas as pd
import NN

#Reading data
data = pd.read_csv("train_data.csv", header = None)
labels = pd.read_csv("train_labels.csv",  header = None)


#Dividing into 90% training and 10% testing test
train_count = np.random.choice(data.index, size=int(len(data)*0.9), replace=False)
train_data, test_data = data.loc[train_count], data.drop(train_count)
train_labels, test_labels = labels.loc[train_count], labels.drop(train_count)


##Further splitting train data into train and validation set in 80% and 20%
train_count_new = np.random.choice(train_data.index, size=int(len(train_data)*0.8), replace=False)
train_data_new, validation_data = train_data.loc[train_count_new], train_data.drop(train_count_new)
train_labels_new, validation_labels = train_labels.loc[train_count_new], train_labels.drop(train_count_new)
#train_data_new

##Standardizing training data and validation data
X_train = (train_data_new - train_data_new.mean()) / (train_data_new.std())
X_validate = (validation_data - validation_data.mean()) / (validation_data.std())
X_test = (test_data - test_data.mean()) / (test_data.std())

##Replacing NaN values with zero
X_train = X_train.fillna(0)
X_validate = X_validate.fillna(0)
X_test = X_test.fillna(0)

##Getting labels of three sets in numpy format from pandas dataframe
y_train = train_labels_new.to_numpy() 
y_validate = validation_labels.to_numpy()
y_test = test_labels.to_numpy()

```

```python
##Initializing and getting the neural network
NN = NN.NeuralNetwork(n_input = 784, n_hidden = 300, n_output = 4,learning_rate=0.001)
```

```python
##Training neural network by specifying parameters
NN.train( train_data_new, y_train, validation_data, y_validate,test_data, y_test, epochs=700, decay=0.0001)
```

```python
#Predicting labels in hot encoded format
predicted = NN.predict(X_test)
one_hot_predictions = np.empty((0,4), int)
#one_hot_predictions.shape
for i in range(len(predicted)):
    row = (predicted[i] == np.amax(predicted[i])).astype(float).reshape(1,4)
    one_hot_predictions = np.append(one_hot_predictions, row, axis=0)
#print(one_hot_predictions)

##Getting accuracy of the trained model
percent = np.sum((one_hot_predictions  == y_test).all(1)) / len(y_test) * 100
percent
```

```python
def accuracy(y_true, y_pred):
    if not (len(y_true) == len(y_pred)):
        print('Size of predicted and true labels not equal.')
        return 0.0

    corr = 0
    for i in range(0,len(y_true)):
        corr += 1 if (y_true[i] == y_pred[i]).all() else 0

    return corr/len(y_true)
```

```python
accuracy(y_test, one_hot_predictions)
```
